```python
import pandas as pd
import os
```

```python

# data from alsaive@institutlyfe.com
data_path = "/home/INT/meunier.d/Data/Data_Cities/Data_wine_ALSaive"
wine_file = os.path.join(data_path, "cleaned_wine_data.csv")

# read csv as DataFrame
df = pd.read_csv(wine_file)
print(df)
```

```python
# keep a subset of the DataFrame 
# the analysis could also be done on the full dataset , 
# in this case, comment the following code and force df_elem = df in the following 
for elem, df_elem in df.groupby(["country", "province"]):
    print(elem)
    if elem == ('France', 'Rhône Valley'):
        print(df_elem)
        break
        
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# computing linear regression between price and points
list1 = df_elem["price"]
list2 = df_elem["points"]

list1 = pd.DataFrame(list1)

model=LinearRegression()
model.fit(list1, list2)

y_pred=model.predict(list1)
```

```python
import matplotlib.pyplot as plt

fig_1=plt.figure(figsize=(8,8))
plt.scatter(list1, list2, color='blue',label='y_test')
plt.plot(list1, y_pred, color='red', label='y_pred')
plt.xlabel('price')
plt.ylabel('points')
plt.legend(loc="upper left")
```

```python
# computing residuals
y_resid =  list2 - y_pred
print(y_resid)

# computing studentized residuals
mean = y_resid.mean()
std = y_resid.std()

std_resid = (y_resid - mean)/std
print(std_resid)
```

```python
# plotting histogram of residuals
fig_1=plt.figure(figsize=(8,8))
plt.hist(std_resid, bins = 90)
```

```python
import numpy as np
outliers_index = np.abs(std_resid.values) > 2
print(np.sum(outliers_index))
list1_out = list1[outliers_index]
list2_out = list2[outliers_index]
print(list1_out)
print(list2_out)
```

```python

#plot outliers in red
fig_1=plt.figure(figsize=(8,8))
plt.scatter(list1, list2, color='blue',label='y_test')
plt.scatter(list1_out, list2_out, color='red', label = 'red')
plt.plot(list1, y_pred, color='red', label='y_pred')
plt.xlabel('price')
plt.ylabel('points')
plt.legend(loc="upper left")
```

```python
# good outliers are the one with residuals 
best_outliers = df_elem[std_resid.values > 2] 
print(best_outliers['title'])
print(best_outliers['price'])
```

```python

#plot good outliers in purple
fig_1=plt.figure(figsize=(8,8))
plt.scatter(list1, list2, color='blue',label='y_test')
plt.scatter(list1_out, list2_out, color='red', label = 'red')

plt.scatter(list1[std_resid.values > 2], list2[std_resid.values > 2], color='purple', label = 'purple')
plt.plot(list1, y_pred, color='red', label='y_pred')
plt.xlabel('price')
plt.ylabel('points')
plt.legend(loc="upper left")
```
